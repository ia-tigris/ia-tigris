```python
odom_df["pose.pose.position.x"] # east
```

```python
odom_df["pose.pose.position.y"] # north
```

```python
new_df = odom_df[["pose.pose.position.x", "pose.pose.position.y", "pose.pose.position.z"]].copy()
new_df
```

```python
new_df["eastings"] = odom_df["pose.pose.position.x"] + easting
new_df["northings"] = odom_df["pose.pose.position.y"] + northing
# new_df["altitude"] = odom_df["pose.pose.position.z"]
new_df.drop(["pose.pose.position.x", "pose.pose.position.y", "pose.pose.position.z"], axis=1, inplace=True)
new_df
```

```python
latlons_df = new_df.apply(lambda x: utm.to_latlon(x["eastings"], x["northings"], zone, letter), axis=1, result_type="expand")
latlons_df
```

```python
latlons_df["altitude"] = odom_df["pose.pose.position.z"]
latlons_df
```

```python
latlons_df.drop_duplicates(keep='last', inplace=True, ignore_index=False)
latlons_df
```

```python
coords = latlons_df.to_numpy()
coords
```

```python
coords.shape
```

```python
import simplekml
```

```python
import os
from os.path import basename, splitext
```

```python
kml = simplekml.Kml()

EVERY_OTHER = 80

linestring = kml.newlinestring()
linestring.coords = coords[::EVERY_OTHER, [1, 0, 2]]
linestring.altitudemode = simplekml.AltitudeMode.relativetoground
linestring.extrude = 1
linestring.style.linestyle.width = 3
linestring.style.linestyle.color = '9900ffff'
file_out = splitext(basename(file_name))[0]
kml.save(file_out + ".kml")
```
