```python
import numpy as np
import matplotlib
import matplotlib.colors
import matplotlib.pyplot as plt
```

```python
def plot_vectorfield(odometry_belief_map):
    width, height, xy = odometry_belief_map.shape
    x_axis = np.arange(0, width, 1)
    y_axis = np.arange(0, width, 1)

    
    x_components = odometry_belief_map[:, :, 0]
    y_components = odometry_belief_map[:, :, 1]
    
    fig, ax = plt.subplots()
    # https://github.com/matplotlib/matplotlib/issues/7909/#issuecomment-274285331
    q = ax.quiver(x_axis, y_axis, x_components, y_components, angles='xy', scale_units='xy', scale=1)

    plt.grid()
    plt.show()

```

```python
odometry_belief_map = np.zeros((10, 10, 2), dtype=float)
```

```python
odometry_belief_map[1, 0, 0] = 1
odometry_belief_map[1, 0, 1] = 1
odometry_belief_map[0, 1, 0] = 1
```

```python
plot_vectorfield(odometry_belief_map)
```
